$$
\begin{align*}
  \begin{cases}
    8.1x_1 + 8.5x_2 &= 72 \\
    -3.2x_1 + 15.2x_2 &= 42 \\
    10.2x_1+8.8x_2 &= 205 \\
  \end{cases}
\end{align*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eq1(x1: float):
    return (72 - 8.1*x1) / 8.5

def eq2(x1: float):
    return (42 + 3.2*x1) / 15.2

def eq3(x1: float):
    return (205 - 10.2*x1) / 8.8
```

```python
x1_values = np.linspace(-10, 20, 400)

eq1_x2_values = eq1(x1_values)
eq2_x2_values = eq2(x1_values)
eq3_x2_values = eq3(x1_values)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x1_values, eq1_x2_values, linestyle='--', color='red')
plt.ylabel('x2')
plt.title(r'$8.1x_1 + 8.5x_2 = 72$')
plt.ylim(-80, 40)

# Plot equation 2
plt.subplot(1, 3, 2)
plt.plot(x1_values, eq2_x2_values, linestyle=':', color='green')
plt.xlabel('x1')
plt.title(r'$-3.2x_1 + 15.2x_2 = 42$')
plt.ylim(-80, 40)

plt.subplot(1, 3, 3)
plt.plot(x1_values, eq3_x2_values, linestyle='-.', color='blue')
plt.title(r'$10.2x_1+8.8x_2 = 205$')
plt.ylim(-80, 40)

plt.tight_layout()
plt.show()
```

```python
eq12L = np.array([[8.1, 8.5], [-3.2, 15.2]])
eq12R = np.array([72, 42])
eq12Inter = np.linalg.solve(eq12L, eq12R)

eq13L = np.array([[8.1, 8.5], [10.2, 8.8]])
eq13R = np.array([72, 205])
eq13Inter = np.linalg.solve(eq13L, eq13R)

eq23L = np.array([[-3.2, 15.2], [10.2, 8.8]])
eq23R = np.array([42, 205])
eq23Inter = np.linalg.solve(eq23L, eq23R)
```

```python
x1_values = np.linspace(-10, 80, 1200)

eq1_x2_values = eq1(x1_values)
eq2_x2_values = eq2(x1_values)
eq3_x2_values = eq3(x1_values)
```

```python
plt.figure(figsize=(16, 8), dpi=150)

plt.plot(x1_values, eq1_x2_values, label='f[0]', linestyle='--', color='red')
plt.text(-8, 7, r'$8.1x_1 + 8.5x_2 = 72$', rotation=-23, fontsize=12, color='black')

plt.plot(x1_values, eq2_x2_values, label='f[1]', linestyle=':', color='green')
plt.text(20, 8, r'$-3.2x_1 + 15.2x_2 = 42$', rotation=5.5, fontsize=12, color='black')

plt.plot(x1_values, eq3_x2_values, label='f[2]', linestyle='-.', color='blue')
plt.text(-8, 20, r'$10.2x_1+8.8x_2 = 205$', rotation=-26, fontsize=12, color='black')

plt.fill_between(x1_values, eq1_x2_values, np.minimum(eq2_x2_values, eq3_x2_values), where=(eq1_x2_values < eq2_x2_values) & (eq1_x2_values < eq3_x2_values), color='lightgrey', alpha=0.5)

plt.title("All lines in one figure")
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True, linestyle='-', linewidth=0.5)

plt.show()
```
